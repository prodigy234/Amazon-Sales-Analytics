```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from IPython.display import display

# Ensure inline plotting for Jupyter Notebook
def enable_inline_plots():
    from IPython import get_ipython
    get_ipython().run_line_magic('matplotlib', 'inline')

enable_inline_plots()

# Load Dataset
file_path = 'amazon.csv'
df = pd.read_csv(file_path)

# Display column names to debug missing columns
print("Available Columns:", df.columns.tolist())

# Basic Analysis
display(df.head())
print("Basic Information:")
print(df.info())
print("\nSummary Statistics:")
display(df.describe())
print("\nMissing Values:")
display(df.isnull().sum())

# Handling missing values
df.dropna(inplace=True)

# Convert rating to numeric if applicable
if 'rating' in df.columns:
    df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Sales Distribution (Using discounted_price as sales metric)
if 'discounted_price' in df.columns:
    print("Generating Sales Visualizations...")
    
    plt.figure(figsize=(10, 5))
    sns.histplot(df['discounted_price'], bins=30, kde=True, color='blue')
    plt.title('Discounted Price Distribution')
    plt.xlabel('Discounted Price')
    plt.ylabel('Frequency')
    plt.show()
    
    # Top 10 Best Selling Products (Assuming higher ratings indicate better selling products)
    if 'product_name' in df.columns:
        top_products = df.groupby('product_name')['rating'].count().nlargest(10)
        
        plt.figure(figsize=(12, 6))
        sns.barplot(y=top_products.index, x=top_products.values, palette='viridis', hue=top_products.values, legend=False)
        plt.title('Top 10 Most Reviewed Products')
        plt.xlabel('Number of Reviews')
        plt.ylabel('Product Name')
        plt.show()

# Customer Review Analysis
if 'review_title' in df.columns:
    title_counts = df['review_title'].value_counts().nlargest(10)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(y=title_counts.index, x=title_counts.values, palette='coolwarm', hue=title_counts.values, legend=False)
    plt.title('Top 10 Most Popular Review Titles')
    plt.xlabel('Count')
    plt.ylabel('Review Title')
    plt.show()

# Word Cloud for Review Content
if 'review_content' in df.columns:
    review_text = ' '.join(df['review_content'].dropna())
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(review_text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Review Content')
    plt.show()

# Distribution of Ratings
if 'rating' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=df['rating'], palette='pastel', hue=df['rating'], legend=False)
    plt.title('Distribution of Ratings')
    plt.xlabel('Rating')
    plt.ylabel('Count')
    plt.show()

# Discount Percentage Analysis
if 'discount_percentage' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df['discount_percentage'], bins=30, kde=True, color='green')
    plt.title('Discount Percentage Distribution')
    plt.xlabel('Discount Percentage')
    plt.ylabel('Frequency')
    plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()

# Category-wise Average Ratings
if 'category' in df.columns and 'rating' in df.columns:
    category_ratings = df.groupby('category')['rating'].mean().nlargest(10)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(y=category_ratings.index, x=category_ratings.values, palette='plasma', hue=category_ratings.index, legend=False)
    plt.title('Top 10 Categories by Average Rating')
    plt.xlabel('Average Rating')
    plt.ylabel('Category')
    plt.show()

print("Data Analysis Complete!")
```
